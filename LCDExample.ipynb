#### Environment Setup

```python
import requests 
from requests import Request, Session
import getpass
import pandas as pd
from io import StringIO
```

```python
def getToken(user, password):
    response = requests.request("GET", "https://mist.morningstar.com/jwt/token", auth=(user, password))
    if response.status_code==401:
        print("Invalid Credentials")
        return
    return response.text

def login():
    user = input('Email: ')
    password = getpass.getpass('Password: ')
    token = getToken(user, password)
    return token
```

#### Get Token

```python
token = login()
```

#### Portfolio API Call

```python
# Edit data to change input parameters
# Must use 'smoketest' for parsing later

client = 'smoketest'

data={
    'feedType': 'CON',
    'client': client,
    'portfolioKey': '2931083',
    'startDate': '2025-09-30',
    'endDate': '2025-10-31',
    'frequency': 'Monthly', #frequency defaults to Daily, supports Monthly and ProformEffective #
    'statementType': 'FIXED_INCOME', #supports EQUITY, PRIVATE_EQUITY, FUND
    'portfolioType': 'Live',
 }

request = Request(
    'POST',
    'https://portfoliocalculation-prd.us-east-1.mif0286.eas.morningstar.com/feeds/v1/generations/download',
    files = {
        'feedType': (None, data['feedType']),
        'client': (None, data['client']),
        'portfolioKey': (None, data['portfolioKey']),
        'startDate': (None, data['startDate']),
        'endDate': (None, data['endDate']),
        'frequency': (None, data['frequency']),
        'statementType': (None, data['statementType']),
        'portfolioType': (None, data['portfolioType'])
    },
    headers = {
        'Authorization': 'Bearer' + token    
    }
).prepare()

s = Session()
response = s.send(request)
port_text = response.text
port_statement = pd.read_csv(StringIO(port_text), sep='\t')
```
